## **Exercice 6 — Calculer la luminosité d'une étoile avec la loi de Planck**

En utilisant les outils appris lors de cette séance, vous allez :

1. **Déterminer la luminosité totale** de votre étoile à l’aide de la **loi de Stefan-Boltzmann (<https://fr.wikipedia.org/wiki/Loi_de_Stefan-Boltzmann>)** :

   $$
   L = 4 \pi R^2 \sigma T^4
   $$

   où :
   - $L$ est la **luminosité totale** (puissance totale rayonnée) en $\mathrm{W}$,
   - $R$ est le **rayon de l’étoile** en $\mathrm{m}$,
   - $\sigma = 5.670374419 \times 10^{-8} \, \mathrm{W.m^{-2}.K^{-4}}$ est la **constante de Stefan-Boltzmann**,
   - $T$ est la **température de surface** de l’étoile, en $\mathrm{K}$.

2. **Calculez la longueur d'onde du pic d'émission à l'aide de la loi de Wien (<https://fr.wikipedia.org/wiki/Loi_du_d%C3%A9placement_de_Wien>)** : 

   $$
   \lambda_\mathrm{max} = \sigma_W / T
   $$

   où :
   - $\lambda_\mathrm{max}$ est la **longueur d'onde** du pic de luminosité en $\mathrm{m}$,
   - $\sigma_W = 2.898e-3 \, \mathrm{m.K^{-1}}$ est la **constante de Wien**,
   - $T$ est la **température de surface** de l’étoile, en $\mathrm{K}$.

3. **Calculer sa luminance spectrale** (ou intensité spectrale) à l’aide de la **loi de Planck (https://fr.wikipedia.org/wiki/Loi_de_Planck)** :

   $$
   B_\lambda(T) = \frac{2 h c^2}{\lambda^5} \cdot \frac{1}{e^{\frac{h c}{\lambda k_B T}} - 1}
   $$

   où :

   - $B_\lambda(T)$ est la **luminance spectrale** du corps noir, en $\mathrm{W.m^{-3}.sr^{-1}}$  
   - $\lambda$ est la **longueur d’onde** du rayonnement, en $\mathrm{m}$  
   - $T$ est la **température absolue** du corps noir, en $\mathrm{K}$  
   - $h = 6.62607015 \times 10^{-34} \, \mathrm{J.s}$ est la **constante de Planck**  
   - $c = 2.99792458 \times 10^8 \, \mathrm{m.s^{-1}}$ est la **vitesse de la lumière**  
   - $k_B = 1.380649 \times 10^{-23} \, \mathrm{J.K^{-1}}$ est la **constante de Boltzmann**

## **Pour aller plus loin**

Il est possible de produire des graphiques interactifs avec la librairie **`ipywidgets`**. En utilisant les widgets tels que les sliders, il est possible de modifier les paramètres d'une fonction et de visualiser en temps réel les changements sur le graphique. Voici un exemple d'implémentation pour la loi de Planck, où il est possible de modifier la température de l'étoile.

En utilisant les widgets `FloatSlider` ou `IntSlider` créez une interface interactive qui permet non seulement d'ajuster la température de l'étoile, mais aussi les limites de longueur d'onde. 

Finalement, utilisez `Checkbox` afin d'afficher ou non la longueur d'onde du pic d'émission (en utilisant la condition `if`, que nous étudierons dans les futures séances).

````python
from ipywidgets import interact, FloatSlider, IntSlider, VBox

def interactive_planck_plot(temperature):
    wavelengths = np.linspace(100e-9, 5000e-9, 1000)
    B_lambda = planck_law(wavelengths, temperature)

    plt.figure(figsize=(8, 4))
    plt.plot(wavelengths, B_lambda, color='darkorange', label="Luminance")
    plt.legend()    
    plt.grid(True)
    plt.title(f"Spectre du corps noir – {nom_etoile} (T = {temperature} K)")
    plt.xlabel("x")
    plt.ylabel("y")
    plt.show()

temperature_slider=FloatSlider(value=1, min=2000, max=10000, step=500, description="temperature")
ui = VBox([temperature_slider]) 

interact(interactive_planck_plot, temperature=temperature_slider) 
````
